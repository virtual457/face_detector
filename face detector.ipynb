```python
import cv2
import matplotlib.pyplot as plt
```

```python
def detect_faces(cascade, test_image, scaleFactor = 1.1):
    # create a copy of the image to prevent any changes to the original one.
    image_copy = test_image.copy()

    #convert the test image to gray scale as opencv face detector expects gray images
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)

    # Applying the haar classifier to detect faces
    faces_rect = cascade.detectMultiScale(gray_image, scaleFactor=scaleFactor, minNeighbors=5)

    for (x, y, w, h) in faces_rect:
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 15)

    return (image_copy,faces_rect)
```

```python
def ret_face(x):
    haar_cascade_face = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_default.xml')
    test_image = cv2.imread(x)
    test_image2 = cv2.imread(x)

      # Converting to grayscale
    test_image_gray = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)
      # Displaying grayscale image
    plt.imshow(test_image_gray, cmap='gray')
    #returns face with mark and co-ordinates
    (faces_rects,blast) = detect_faces(haar_cascade_face, test_image)
     #convert to RGB and display image
    overall=[]
    blue=plt.imread(x)
    for i in range(blast[0][0],blast[0][0]+blast[0][2]):
        o=[]
        for j in range (blast[0][1],blast[0][1]+blast[0][3]):
            o.append(blue[i][j])
        overall.append(o)
    return overall
```
